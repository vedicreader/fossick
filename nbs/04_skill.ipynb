This module is the target of fossick's `pyskills` entry point
(`[project.entry-points.pyskills] fossick = "fossick.skill"` in `pyproject.toml`). A pyskills host
discovers it with `list_pyskills()` -- which reads the first paragraph of this module's docstring via
AST, without importing -- and loads it with `from fossick.skill import *` followed by `doc(fossick.skill)`.

`__all__` is a curated routing surface (star imports are the pyskills convention), one entry per kind
of web task:

| Input | Call |
|---|---|
| search query | `search(q)`; `images`/`news`/`videos`; `google(q)` for real Google ranking (slow) |
| result URL | `extract(url)` or `fetch(url)` |
| static page | `fetch(url)`; `to_md(page, sel='main')` |
| JS-rendered / bot-protected | `fetch(url, heavy=True)` / `fetch(url, stealthy=True)` |
| many links / known URLs | `crawl(url, follow_sel='a[href]')` / `fetch_all(urls)` |
| hidden JSON API | `find_xhr(url, pattern='*api*')` then `paginate_api(...)` |
| arXiv / YouTube / GitHub | `read_arxiv` / `read_yt`, `search_yt`, `download_yt` / `read_gh_file`, `read_gh_repo` |
| paper title to DOI | `lookup_doi(title)` |
| URL or PDF to notebook | `url2nb(url)` / `pdf2nb(url_or_path)` |
| authenticated / interactive | `cdp_connect()` + `pg.ax_tree()` + `pg.fill_text()`/`pg.click_and_wait()`; `syncy(coro)` from sync code |
